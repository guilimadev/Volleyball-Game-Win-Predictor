```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics

import requests

df_final = pd.DataFrame()
```

```python

page = 1239
while page != 1245:
  url = f'https://cbv-web.dataproject.com/MatchStatistics.aspx?mID={page}&ID=13'

  res = requests.get(url)
  html_doc = res.text
  soup2 = BeautifulSoup(html_doc, 'html.parser')
  
  home_sets = int(soup2.find('span', id='Content_Main_LBL_WonSetHome').text)
  away_sets = int(soup2.find('span', id='Content_Main_LBL_WonSetGuest').text)




  playername = []
  for pn in soup2.find_all("span", id="PlayerName"):
    playername.append(pn.text)

  arr_name = np.array(playername)


  x = np.where(arr_name == 'TOTAIS')

  team1 = arr_name[:x[0][0]]

  team2 = arr_name[x[0][0]+1:x[0][1]]

  
  spike = []
  for pn in soup2.find_all("span", id="SpikeWin"):
    spike.append(pn.text)

  spike_porc = []
  for pn in soup2.find_all("span", id=["SpikePos", "SpikePos0"]):
    spike_porc.append(pn.text)

  spike_error = []
  for pn in soup2.find_all("span", id=["SpikeErr"]):
    spike_error.append(pn.text)

  porc_recp = []
  for pn in soup2.find_all("span", id=["RecPos", "RecPos0"]):
    porc_recp.append(pn.text)

  total_atq = []
  for pn in soup2.find_all("span", id=["SpikeTot"]):
    total_atq.append(pn.text)

  total_blk = []
  for pn in soup2.find_all("span", id=["BlockWin"]):
    total_blk.append(pn.text)

  error_serve = []
  for pn in soup2.find_all("span", id=["ServeErr"]):
    error_serve.append(pn.text)



  result = home_sets - away_sets
 
  if result < 0:
    home_vencedor = False
  else:
    home_vencedor = True


  n_sets = home_sets + away_sets

  
  dict_game = {  
    'id_jogo' : page,
    'team home': soup2.find('span', id='Content_Main_LBL_HomeTeam').text,
    'total atq home' : total_atq[x[0][0]],
    'pontos ataques home' : spike[x[0][0]],
    'atq error home' : spike_error[x[0][0]],
    '% atq home' : spike_porc[x[0][0]], 
    '% rec home' : porc_recp[x[0][0]],
    'bloqueio home' : total_blk[x[0][0]] ,
    'server error home' : error_serve[x[0][0]],
    'team away': soup2.find('span', id='Content_Main_LBL_GuestTeam').text,
    'total atq away' : total_atq[x[0][1]],
    'pontos ataques away' : spike[x[0][1]], 
    'atq error away' : spike_error[x[0][1]],
    '% atq away' : spike_porc[x[0][1]],
    '% rec away' : porc_recp[x[0][1]], 
    'bloqueio away' : total_blk[x[0][1]] , 
    'server error away' : error_serve[x[0][1]],
    'qtd sets': n_sets,
    'vencedor' :  home_vencedor    
  }



  page = page + 1
  df_final = df_final.append([dict_game])

page = 1251
while page != 1310:
  url = f'https://cbv-web.dataproject.com/MatchStatistics.aspx?mID={page}&ID=13'

  res = requests.get(url)
  html_doc = res.text
  soup2 = BeautifulSoup(html_doc, 'html.parser')
  
  home_sets = int(soup2.find('span', id='Content_Main_LBL_WonSetHome').text)
  away_sets = int(soup2.find('span', id='Content_Main_LBL_WonSetGuest').text)




  playername = []
  for pn in soup2.find_all("span", id="PlayerName"):
    playername.append(pn.text)

  arr_name = np.array(playername)


  x = np.where(arr_name == 'TOTAIS')

  team1 = arr_name[:x[0][0]]

  team2 = arr_name[x[0][0]+1:x[0][1]]

  spike = []
  for pn in soup2.find_all("span", id="SpikeWin"):
    spike.append(pn.text)

  spike_porc = []
  for pn in soup2.find_all("span", id=["SpikePos", "SpikePos0"]):
    spike_porc.append(pn.text)

  spike_error = []
  for pn in soup2.find_all("span", id=["SpikeErr"]):
    spike_error.append(pn.text)

  porc_recp = []
  for pn in soup2.find_all("span", id=["RecPos", "RecPos0"]):
    porc_recp.append(pn.text)

  total_atq = []
  for pn in soup2.find_all("span", id=["SpikeTot"]):
    total_atq.append(pn.text)

  total_blk = []
  for pn in soup2.find_all("span", id=["BlockWin"]):
    total_blk.append(pn.text)

  error_serve = []
  for pn in soup2.find_all("span", id=["ServeErr"]):
    error_serve.append(pn.text)


  result = home_sets - away_sets
 
  if result < 0:
    home_vencedor = False
  else:
    home_vencedor = True
  
  n_sets = home_sets + away_sets


  dict_game = {  
    'id_jogo' : page,
    'team home': soup2.find('span', id='Content_Main_LBL_HomeTeam').text,
    'total atq home' : total_atq[x[0][0]],
    'pontos ataques home' : spike[x[0][0]],
    'atq error home' : spike_error[x[0][0]],
    '% atq home' : spike_porc[x[0][0]], 
    '% rec home' : porc_recp[x[0][0]],
    'bloqueio home' : total_blk[x[0][0]] ,
    'server error home' : error_serve[x[0][0]],
    'team away': soup2.find('span', id='Content_Main_LBL_GuestTeam').text,
    'total atq away' : total_atq[x[0][1]],
    'pontos ataques away' : spike[x[0][1]], 
    'atq error away' : spike_error[x[0][1]],
    '% atq away' : spike_porc[x[0][1]],
    '% rec away' : porc_recp[x[0][1]], 
    'bloqueio away' : total_blk[x[0][1]] , 
    'server error away' : error_serve[x[0][1]],
    'qtd sets': n_sets,
    'vencedor' :  home_vencedor    
  }



  page = page + 1
  df_final = df_final.append([dict_game])


```

```python

df_final['% atq home'] = df_final['% atq home'].str.rstrip('%').astype('int')
df_final['% rec home'] = df_final['% rec home'].str.rstrip('%').astype('int')
df_final['% atq away'] = df_final['% atq away'].str.rstrip('%').astype('int')
df_final['% rec away'] = df_final['% rec away'].str.rstrip('%').astype('int')
df_final = df_final.set_index('id_jogo')
```
